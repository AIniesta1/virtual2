```python
import numpy as np
from sklearn.metrics import root_mean_squared_error
```

1-1- Crea dos vectores con 10 valores aleatorios int entre 30 y 60 (vector 1) y con 10 valores float del 0.1 al 9.9 (vector 2) 
2-El primer vector se corresponderá con el valor real y el segundo vector con el valor predicho por el modelo
3-Calcula el RMSE con la función root_mean_squared_error de Scikit Learn y además también utilizando la forma tradicional
4- Imprime el resultado de ambas formas.
5- ¿Hay diferencia entre los resultados?

```python
v_real = np.random.randint(30,60,10)
print(v_real)
v_predicho = np.random.uniform(0.1,9.9,10)
print(v_predicho)
rmse_sklearn = root_mean_squared_error(v_real, v_predicho)  
print("RMSE con Scikit-learn:", rmse_sklearn)
```

```python
errors = (v_real - v_predicho) ** 2
mse = np.mean(errors)
rmse = np.sqrt(mse) 
print(f"RMSE: {rmse}")
```

### **Contexto:**

El RMSE es una métrica común en Machine Learning para evaluar la precisión de un modelo.

### **Tareas:**

1. Define dos arrays: uno con valores reales (`y`) y otro con valores predichos (`y_pred`).
2. Calcula el error cuadrático medio (MSE) elevando al cuadrado las diferencias entre `y` y `y_pred`, promediando los valores y tomando la raíz cuadrada.
3. Imprime el resultado del RMSE.
4. Repite el cálculo utilizando la función `root_mean_squared_error` de `sklearn.metrics`.

```python
y = np.random.randint(1,10,10)
print(y)
y_pred = np.random.uniform(0.1,9.9,10)
print(y_pred)
errors = (y - y_pred) ** 2
mse = np.mean(errors)
rmse = np.sqrt(mse) 
print(f"RMSE: {rmse}")
rmse_sklearn = root_mean_squared_error(y, y_pred)  
print("RMSE con Scikit-learn:", rmse_sklearn)
```

Softmax convierte logits en probabilidades normalizadas para clasificación en modelos de Machine Learning.

### **Tareas:**

1. Define un array con tres valores representando logits de un modelo.
2. Aplica la función exponencial a cada valor y normaliza dividiendo entre la suma total.
3. Imprime las probabilidades resultantes.

```python
logits = np.array([2.0, 1.0, 0.1])
exp = np.exp(logits)
norm = exp / np.sum(exp)
print(norm) 
```

```python
from sklearn.utils.extmath import softmax

probabilities_sklearn = softmax(logits.reshape(1, -1))[0]
print("Probabilidades softmax con Scikit-learn:", probabilities_sklearn)
```

```python
from scipy.special import softmax

probabilities = softmax(logits)
print("Probabilidades softmax con Spicy:", probabilities)
```

```python
import pandas as pd
import numpy as np

# Crear un DataFrame con valores aleatorios
df = pd.DataFrame({
    "id": range(1, 11),
    "ventas": np.random.randint(100, 1000, size=10)  # Números aleatorios entre 100 y 1000
})

# Agregar una columna con la media de 'ventas'
df["media_ventas"] = df["ventas"].mean()

print(df)

# CALCULAR LA MEDIANA

df["mediana_ventas"] = df["ventas"].median() # .median() devuelve el valor central de la distribución.
print(df)


# CALCULAR LA DESVIACIÓN ESTÁNDAR

df["desviacion_ventas"] = df["ventas"].std() #.std() calcula la dispersión de los datos respecto a la media.
print(df)

# NORMALIZAR VALORES CON MEDIA Y DESVIACIÓN ESTÁNDAR 

df["ventas_normalizadas"] = (df["ventas"] - df["ventas"].mean()) / df["ventas"].std() # Se aplica la fórmula de normalización Z-score: (X - media) / desviación estándar
print(df)

df["media_v_norm"] = df["ventas_normalizadas"].median()
df["desviacion_v_norm"] = df["ventas_normalizadas"].std()
df
```
