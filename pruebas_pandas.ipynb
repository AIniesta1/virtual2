```python
import seaborn as sns
import pandas as pd 

# Cargar el dataset Titanic
titanic = sns.load_dataset('titanic')

# Seleccionar la columna 'Age' como Serie
age_series = titanic['age']

# Mostrar las primeras filas de la Serie 'age'
titanic
```

```python
print(age_series[0])
print(age_series[890])
print(titanic.iat[0,3])
print(titanic.iat[-1,3])
print(age_series[10])
```

```python
print(age_series + 5)
grouped = titanic.groupby("sex")["age"]
for name, group in grouped:
    if name == "female":
        print(group * 2)

```

```python
total = age_series.sum()
mean = age_series.mean()
median = age_series.median()
minimum = age_series.min()
maximum = age_series.max()
std_dev = age_series.std()
print(total)
print(mean)
print(median)
print(minimum)
print(maximum)
print(std_dev)
```

```python
age_series[0] = 100
print(age_series)
nueva_serie = pd.Series([60,70,80])
print(pd.concat([age_series, nueva_serie], ignore_index= True))
```

```python
titanic["age fem 2"] = titanic["age"].where(titanic["sex"] != "female", other=titanic["age"] * 2)
titanic
```

```python
titanic['edad'] = titanic.groupby('sex')['age'].transform(lambda x: x * 2 if x.name == 'female' else x)
titanic
```

```python
import numpy as np 
titanic["durum mixto"] = np.random.choice(["algo",1,True,False,2,3,4,5,6,7,8,9], size=len(titanic))
titanic["durum mixto"] = pd.to_numeric(titanic["durum mixto"], errors="coerce")
titanic
```
